```python
new_data=new_data.reset_index()
```

```python
drop_col=['price_range','currency','index','zipcode']
new_data.drop(columns=drop_col,axis=1,inplace=True)
```

```python
new_data['establishment'].unique()
```

```python
# Working on null values
null_values=pd.isnull(new_data['establishment'])
null_data=new_data[null_values]
lst=list(null_data['res_id'])
```

```python

def fill_data(x):
    if pd.isna(x)==True:
        return "Restaurant/Shop"
    else:
        return x    
```

```python
new_data['establishment']=new_data['establishment'].apply(fill_data)
new_data
```

```python
# Working on null values
null_values=pd.isnull(new_data['establishment'])
null_data=new_data[null_values]
```

```python
def high(x):
    if pd.isna(x)==True:
        return "Available, Cash"
    else:
        return x    
```

```python
new_data['highlights']=new_data['highlights'].apply(high)
```

```python
new_data.isnull().sum()
```

```python
def time(x):
    if pd.isna(x)==True:
        return "11 AM to 11 PM"
    else:
        return x
new_data['timings']=new_data['timings'].apply(time)        
```

```python
def food(x):
    if pd.isna(x)==True:
        return "Fast Food"
    else:
        return x    
```

```python
new_data['cuisines']=new_data['cuisines'].apply(food)
```
